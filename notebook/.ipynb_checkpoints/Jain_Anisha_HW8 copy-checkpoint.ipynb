#### (i) Supervised Learning

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = LinearSVC(penalty='l1', dual=False, max_iter=5000)
param_grid = {'C': np.logspace(-3, 3, 10)}
cross_val = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
num_simulations = 30
metrics_summary = []

for i in range(num_simulations):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=i)
    
    grid_search = GridSearchCV(model, param_grid, cv=cross_val, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_classifier = grid_search.best_estimator_
    
    y_train_pred = best_classifier.predict(X_train)
    y_test_pred = best_classifier.predict(X_test)
    
    scores_train = {
        'Accuracy': accuracy_score(y_train, y_train_pred),
        'Precision': precision_score(y_train, y_train_pred),
        'Recall': recall_score(y_train, y_train_pred),
        'F1 Score': f1_score(y_train, y_train_pred),
        'AUC': roc_auc_score(y_train, y_train_pred)
    }
    scores_test = {
        'Accuracy': accuracy_score(y_test, y_test_pred),
        'Precision': precision_score(y_test, y_test_pred),
        'Recall': recall_score(y_test, y_test_pred),
        'F1 Score': f1_score(y_test, y_test_pred),
        'AUC': roc_auc_score(y_test, y_test_pred)
    }
    metrics_summary.append({'Train': scores_train, 'Test': scores_test})

    print(f"Iteration {i+1}/{num_simulations} ")
    print(f"Train Accuracy: {scores_train['Accuracy']:.3f} | Test Accuracy: {scores_test['Accuracy']:.3f}")

average_metrics = {'Train': {}, 'Test': {}}
for key in scores_train:
    average_metrics['Train'][key] = np.mean([m['Train'][key] for m in metrics_summary])
    average_metrics['Test'][key] = np.mean([m['Test'][key] for m in metrics_summary])

print("-" * 60)
print("Average Metrics Over All Simulations:")
for metric, values in average_metrics.items():
    print(f"{metric} -", values)
print("-" * 60)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
fpr, tpr, _ = roc_curve(y_train, grid_search.decision_function(X_train))
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'ROC curve Train (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Training Data')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
fpr, tpr, _ = roc_curve(y_test, grid_search.decision_function(X_test))
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'ROC curve Test (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Testing Data')
plt.legend(loc='lower right')
plt.show()

cm_train = confusion_matrix(y_train, y_train_pred)
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train)
disp_train.plot()
plt.title('Confusion Matrix for Training Data')
plt.show()

cm_test = confusion_matrix(y_test, y_test_pred)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test)
disp_test.plot()
plt.title('Confusion Matrix for Testing Data')
plt.show()
```

#### (ii) Semi-Supervised Learning/ Self-training (A) (B)

```python
num_runs = 30
test_size = 0.2
random_state = 42
all_metrics = []

for run in range(num_runs):
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=test_size, stratify=y, random_state=run)
    X_train_df = pd.DataFrame(X_train)
    y_train_df = pd.Series(y_train).reset_index(drop=True)
    
    # Selecting labeled and unlabeled data
    pos_indices = y_train_df[y_train_df == 1].index
    neg_indices = y_train_df[y_train_df == 0].index
    labeled_pos_indices = np.random.choice(pos_indices, size=len(pos_indices)//2, replace=False)
    labeled_neg_indices = np.random.choice(neg_indices, size=len(neg_indices)//2, replace=False)
    labeled_indices = np.concatenate([labeled_pos_indices, labeled_neg_indices])
    
    unlabeled_indices = list(set(y_train_df.index) - set(labeled_indices))
    X_labeled = X_train_df.iloc[labeled_indices].values
    y_labeled = y_train_df.iloc[labeled_indices]
    X_unlabeled = X_train_df.iloc[unlabeled_indices].values
    
    # Model and parameter selection
    model = LinearSVC(penalty='l1', dual=False, max_iter=5000)
    param_grid = {'C': np.logspace(-3, 3, 10)}
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_labeled, y_labeled)
    model = grid_search.best_estimator_
    
    # Self-training loop
    while len(X_unlabeled) > 0:
        decision_function = model.decision_function(X_unlabeled)
        farthest_index = np.argmax(np.abs(decision_function))
        X_newly_labeled = X_unlabeled[farthest_index:farthest_index+1]
        y_newly_labeled = model.predict(X_newly_labeled)
        
        # Add newly labeled data to labeled set
        X_labeled = np.vstack([X_labeled, X_newly_labeled])
        y_labeled = np.concatenate([y_labeled, y_newly_labeled])
        
        # Remove from unlabeled set
        X_unlabeled = np.delete(X_unlabeled, farthest_index, axis=0)
        
        # Retrain model
        model.fit(X_labeled, y_labeled)
    
    # Evaluate model
    y_pred_train = model.predict(X_train_df.values)
    y_pred_test = model.predict(X_test)
    scores_train = {
        'Accuracy': accuracy_score(y_train, y_pred_train),
        'Precision': precision_score(y_train, y_pred_train),
        'Recall': recall_score(y_train, y_pred_train),
        'F1 Score': f1_score(y_train, y_pred_train),
        'AUC': roc_auc_score(y_train, y_pred_train)
    }
    scores_test = {
        'Accuracy': accuracy_score(y_test, y_pred_test),
        'Precision': precision_score(y_test, y_pred_test),
        'Recall': recall_score(y_test, y_pred_test),
        'F1 Score': f1_score(y_test, y_pred_test),
        'AUC': roc_auc_score(y_test, y_pred_test)
    }
    all_metrics.append({'Train': scores_train, 'Test': scores_test})
    print(f"Iteration {run+1}/{num_simulations}")
    print(f"Train Accuracy: {scores_train['Accuracy']:.3f} | Test Accuracy: {scores_test['Accuracy']:.3f}")

average_metrics = {'Train': {}, 'Test': {}}
for key in scores_train:
    average_metrics['Train'][key] = np.mean([m['Train'][key] for m in metrics_summary])
    average_metrics['Test'][key] = np.mean([m['Test'][key] for m in metrics_summary])

print("-" * 60)
print("Average Metrics Over All Simulations:")
for metric, values in average_metrics.items():
    print(f"{metric} -", values)
print("-" * 60)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
fpr, tpr, _ = roc_curve(y_train, grid_search.decision_function(X_train))
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'ROC curve Train (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Training Data')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
fpr, tpr, _ = roc_curve(y_test, grid_search.decision_function(X_test))
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label=f'ROC curve Test (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Testing Data')
plt.legend(loc='lower right')
plt.show()

cm_train = confusion_matrix(y_train, y_train_pred)
disp_train = ConfusionMatrixDisplay(confusion_matrix=cm_train)
disp_train.plot()
plt.title('Confusion Matrix for Training Data')
plt.show()

cm_test = confusion_matrix(y_test, y_test_pred)
disp_test = ConfusionMatrixDisplay(confusion_matrix=cm_test)
disp_test.plot()
plt.title('Confusion Matrix for Testing Data')
plt.show()
```

```python
def get_distance(centers, X, cluster_label):
    distances = np.zeros(X.shape[0])
    probabilities = np.zeros((X.shape[0], 2))
    for i, x in enumerate(X):
        center = centers[cluster_label[i]]
        distances[i] = np.linalg.norm(x - center)
        probabilities[i, 0] = np.linalg.norm(x - centers[0])
        probabilities[i, 1] = np.linalg.norm(x - centers[1])
    
    probabilities = softmax(probabilities, axis=1).astype(np.float64)
    return distances, probabilities

def get_majority(X, cluster_label, distances, Y):
    label_to_cluster = np.zeros(2, dtype=int)
    majority_labels = cluster_label.copy()
    for cluster in [0, 1]:
        sample_indices = np.where(cluster_label == cluster)[0]
        top30_indices = sorted(sample_indices.tolist(), key=lambda i: distances[i])[:30]
        majority = Y.iloc[top30_indices].value_counts().index[0]
        majority_labels[sample_indices] = majority
        label_to_cluster[majority] = cluster
    return label_to_cluster, majority_labels

def evaluate_metrics(ground_truth, predict, prob, report):
    acc = report['accuracy']
    fpr, tpr, thresholds = roc_curve(ground_truth, prob, pos_label=1)
    roc_auc = round(auc(fpr, tpr), 4)
    return {
        'accuracy': acc,
        'precision': report['weighted avg']['precision'],
        'recall': report['weighted avg']['recall'],
        'f1-score': report['weighted avg']['f1-score'],
        'auc': roc_auc
    }

def split_data(data, random_state, test_size=0.2):
    train_data, test_data = train_test_split(data, test_size=test_size, stratify=data['label'], random_state=random_state)
    train_X = train_data.iloc[:, 2:].copy()
    train_Y = train_data['label'].copy()
    test_X = test_data.iloc[:, 2:].copy()
    test_Y = test_data['label'].copy()
    return train_X, train_Y, test_X, test_Y

def normalize_data(train_X, test_X):
    scaler = Normalizer()
    normalized_train_X = scaler.fit_transform(train_X)
    normalized_test_X = scaler.transform(test_X)
    return normalized_train_X, normalized_test_X

def monte_carlo_clustering(data, num_runs, algorithm):
    print('=' * 60)
    print(f'Running method for {algorithm}')
    print('=' * 60)
    
    cols = ['accuracy', 'precision', 'recall', 'f1-score', 'auc']
    train_results = pd.DataFrame(columns=cols)
    test_results = pd.DataFrame(columns=cols)
    
    for i in range(num_runs):
        train_X, train_Y, test_X, test_Y = split_data(data, random_state=i)
        train_X, test_X = normalize_data(train_X, test_X)
        
        if algorithm == 'Unsupervised-KMeans':
            clusterer = KMeans(n_clusters=2, n_init=15, random_state=i).fit(train_X)
            train_cluster_label = clusterer.predict(train_X)
            distances, train_pred_prob = get_distance(clusterer.cluster_centers_, train_X, train_cluster_label)
            label_to_cluster, train_pred_Y = get_majority(train_X, train_cluster_label, distances, train_Y)
        elif algorithm == 'Unsupervised-Spectral':
            cluster_label = SpectralClustering(n_clusters=2, assign_labels='kmeans',
                                               affinity='rbf', eigen_tol=1e-2, eigen_solver='arpack',
                                               random_state=0).fit_predict(train_X)
            cluster_centers = [np.mean(train_X[cluster_label == clus], axis=0).tolist() for clus in [0, 1]]
            knn = KNeighborsClassifier(n_neighbors=1)
            knn.fit(train_X, cluster_label)
            train_cluster_label = knn.predict(train_X)
            distances, train_pred_prob = get_distance(cluster_centers, train_X, train_cluster_label)
            label_to_cluster, train_pred_Y = get_majority(train_X, train_cluster_label, distances, train_Y)
        
        pos_cluster = label_to_cluster[0]
        train_pred_prob = train_pred_prob[:, pos_cluster]
        train_report = classification_report(train_Y, train_pred_Y, output_dict=True)
        tmp_train_res = evaluate_metrics(train_Y, train_pred_Y, train_pred_prob, train_report)
        train_results = pd.concat([train_results, pd.DataFrame([tmp_train_res])], ignore_index=True)
        
        if algorithm == 'Unsupervised-KMeans':
            test_cluster_label = clusterer.predict(test_X)
            distances, test_pred_prob = get_distance(clusterer.cluster_centers_, test_X, test_cluster_label)
        elif algorithm == 'Unsupervised-Spectral':
            test_cluster_label = knn.predict(test_X)
            distances, test_pred_prob = get_distance(cluster_centers, test_X, test_cluster_label)
        
        label_to_cluster, test_pred_Y = get_majority(test_X, test_cluster_label, distances, test_Y)
        pos_cluster = label_to_cluster[0]
        test_pred_prob = test_pred_prob[:, pos_cluster]
        test_report = classification_report(test_Y, test_pred_Y, output_dict=True)
        tmp_test_res = evaluate_metrics(test_Y, test_pred_Y, test_pred_prob, test_report)
        test_results = pd.concat([test_results, pd.DataFrame([tmp_test_res])], ignore_index=True)
        
        train_accuracy = train_report['accuracy']
        test_accuracy = test_report['accuracy']
        print(f'Iteration: {i + 1} | Training Accuracy: {round(train_accuracy, 4)} | Testing Accuracy: {round(test_accuracy, 4)}')
    
    print('-' * 80)
    print('\nDetail report of 30th run of Monte-Carlo\n')
    
    calculate_model_summary(train_Y, train_pred_Y, train_pred_prob, 'Training')
    calculate_model_summary(test_Y, test_pred_Y, test_pred_prob, 'Test')
    
    summary = summarize_results(train_results, test_results)
    return summary

algorithm = 'Unsupervised-KMeans'
summary['Unsupervised KMeans'] = monte_carlo_clustering(df, 30, algorithm)
df = pd.DataFrame(summary['Unsupervised KMeans'])
df
```

```python
X = data.iloc[:, 2:]
y = data['Diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

gamma = 1
target_ratio = y.value_counts(normalize=True)

best_gamma = gamma
best_balance_diff = float('inf')

for gamma in np.logspace(-2, 1, 20): 
    clustering = SpectralClustering(n_clusters=2, affinity='rbf', gamma=gamma, random_state=42, assign_labels='discretize')
    labels = clustering.fit_predict(X_scaled)
    cluster_ratio = np.bincount(labels) / len(labels)
    
    balance_diff = abs(cluster_ratio[0] - target_ratio[0]) + abs(cluster_ratio[1] - target_ratio[1])
    
    if balance_diff < best_balance_diff:
        best_balance_diff = balance_diff
        best_gamma = gamma

spectral = SpectralClustering(n_clusters=2, affinity='rbf', gamma=best_gamma, random_state=42, assign_labels='discretize')
labels = spectral.fit_predict(X_scaled)

accuracy = accuracy_score(y, labels)
precision = precision_score(y, labels)
recall = recall_score(y, labels)
f1 = f1_score(y, labels)
auc = roc_auc_score(y, labels)

print(f"Spectral Clustering with gamma={best_gamma} Metrics:")
print(f"Accuracy: {accuracy}, Precision: {precision}, Recall: {recall}, F1: {f1}, AUC: {auc}")
```

## 2. Active Learning Using Support Vector Machines

### (a) Download the banknote authentication Data Set

```python
banknote_data_path = "../data/data_banknote_authentication.txt"

# Read the dataset into a DataFrame
banknote_data = pd.read_csv(banknote_data_path, header=None)

# Assign meaningful column names
banknote_data.columns = ['variance', 'skewness', 'curtosis', 'entropy', 'class']

# Display the dataset
print(banknote_data)

# Display the count of each class
print(banknote_data['class'].value_counts())

# Prepare the feature matrix X and the target vector y
X = banknote_data.drop('class', axis=1)
y = banknote_data['class']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=472/(900+472), random_state=42)
```

### (b) Repeat each of the following two procedures 50 times

#### (i) Passive

```python
pool_size = 10
pool_errors = []

# Passive learning loop
for i in range(10, 901, 10):
    # Select data points for training
    X_train_pool = X_train[:i]
    y_train_pool = y_train[:i]
    
    # Define SVM model with linear kernel
    svm_model = SVC(kernel='linear')
    
    # Define parameter grid for GridSearchCV
    param_grid = {'C': np.logspace(-3, 3, 7)}
    
    # Perform grid search with 5-fold cross-validation
    grid_search = GridSearchCV(svm_model, param_grid, cv=5)
    grid_search.fit(X_train_pool, y_train_pool)
    
    # Get the best SVM model
    best_svm_model = grid_search.best_estimator_
    
    # Calculate test error
    y_pred = best_svm_model.predict(X_test)
    test_error = 1 - accuracy_score(y_test, y_pred)
    
    # Append test error to list
    pool_errors.append(test_error)
    
    # Print test error for this iteration
    print(f"SVM {i // 10}: Test Error = {test_error:.4f}")

# Display test errors for each SVM
for i, error in enumerate(pool_errors, 1):
    print(f"SVM {i}: Test Error = {error:.4f}")
```

#### (ii) Active

```python
pool_size = 10
pool_errors = []

# Active learning loop
for i in range(10, 901, 10):
    # Initialize pool and remaining data
    X_train_pool = X_train[:pool_size]
    y_train_pool = y_train[:pool_size]
    remaining_X_train = X_train[pool_size:i]
    remaining_y_train = y_train[pool_size:i]
    
    while remaining_X_train.shape[0] > 0:
        # Train SVM with current pool
        svm_model = SVC(kernel='linear')
        param_grid = {'C': np.logspace(-3, 3, 7)}
        grid_search = GridSearchCV(svm_model, param_grid, cv=5)
        grid_search.fit(X_train_pool, y_train_pool)
        best_svm_model = grid_search.best_estimator_
        
        # Find the 10 closest data points to the hyperplane
        distances = best_svm_model.decision_function(remaining_X_train)
        closest_indices = np.argsort(np.abs(distances))[:10]
        
        # Add closest data points to the pool
        X_train_pool = np.vstack([X_train_pool, remaining_X_train[closest_indices]])
        y_train_pool = np.concatenate([y_train_pool, remaining_y_train[closest_indices]])
        
        # Remove closest data points from the remaining data
        remaining_X_train = np.delete(remaining_X_train, closest_indices, axis=0)
        remaining_y_train = np.delete(remaining_y_train, closest_indices)
    
    # Train SVM with the final pool
    svm_model = SVC(kernel='linear')
    param_grid = {'C': np.logspace(-3, 3, 7)}
    grid_search = GridSearchCV(svm_model, param_grid, cv=5)
    grid_search.fit(X_train_pool, y_train_pool)
    best_svm_model = grid_search.best_estimator_
    
    # Evaluate on test set
    y_pred = best_svm_model.predict(X_test)
    test_error = 1 - accuracy_score(y_test, y_pred)
    
    # Store test error
    pool_errors.append(test_error)

# Display test errors for each SVM
for i, error in enumerate(pool_errors, 1):
    print(f"SVM {i}: Test Error = {error:.4f}")
```
